```python
import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
# Load YOLOv5 pre-trained model
model = YOLO("yolov5s.pt")
```

```python
# Path to the uploaded image
image_path = "/content/drive/MyDrive/DL/cars-city-traffic-daylight.jpg"
```

```python
# Load image
img = cv2.imread(image_path)
```

```python
# Perform object detection
results = model(image_path)
```

```python
# Get detection output with bounding boxes
detected_img = results[0].plot()  # Draw detections
```

```python
# Convert BGR to RGB for correct visualization
detected_img_rgb = cv2.cvtColor(detected_img, cv2.COLOR_BGR2RGB)
```

```python
# Display original and detected images side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Original Image
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original Image")
ax[0].axis("off")

# Object Detection Result
ax[1].imshow(detected_img_rgb)
ax[1].set_title("Object Detection Output")
ax[1].axis("off")

plt.show()
```
